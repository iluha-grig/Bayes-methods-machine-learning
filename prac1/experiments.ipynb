## Задание 4

```python
p = np.linspace(0, 1, 1000)
```

```python
p.shape[0]
```

```python
p_matrix = np.repeat(p[np.newaxis, :], 1000, 0)
```

```python
def var(prob, val):
    return np.sum(prob * (val - np.sum(prob * val)) ** 2)
```

```python
points_model1 = []
for p1 in p:
    for p2 in p:
        params['p1'] = p1
        params['p2'] = p2
        prob_c_a, val_c = pc_a(np.array([83]), params, 1)
        prob_c_b, val_c = pc_b(np.array([550]), params, 1)
        if var(prob_c_b, val_c) < var(prob_c_a, val_c):
            points_model1.append([p1, p2])
```

```python
len(points_model1) / 1000000
```

```python
points1 = np.array(points_model1)
```

```python
points_model2 = []
for p1 in p:
    for p2 in p:
        params['p1'] = p1
        params['p2'] = p2
        prob_c_a, val_c = pc_a(np.array([83]), params, 2)
        prob_c_b, val_c = pc_b(np.array([550]), params, 2)
        if var(prob_c_b, val_c) < var(prob_c_a, val_c):
            points_model2.append([p1, p2])
```

```python
len(points_model2) / 1000000
```

```python
points2 = np.array(points_model2)
```

```python
fig = plt.figure(figsize=(20, 10))
plt.scatter(p_matrix.ravel(), p_matrix.ravel(order='F'), c='black', s=0.01)
plt.title('All points', fontsize=20)
plt.xlabel('p1', fontsize=16)
plt.ylabel('p2', fontsize=16)
plt.show()
```
